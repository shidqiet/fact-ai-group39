```python
#%load_ext autoreload
#%autoreload 2

from language import Transformer
from sae import Tracerfw, Visualizerfw
from sae.functions import compute_truncated_eigenvalues

import plotly.express as px

import torch

device = "cpu"
```

```python
torch.set_grad_enabled(False)
model = Transformer.from_pretrained("Julianvn/facts-fw-small").to(device)
```

```python
model.generate("Hello world, ", max_length=30)
```

We then instantiate some useful helper objects:
- ``Tracer`` loads SAEs around the MLP of a given layer and contains some helper functions to compute interaction matrices between the two.
- ``Visualizer`` shows the (pre-computed) top activations of SAE features to understand their meaning.

```python
from datasets import load_dataset

dataset = load_dataset(f"tdooms/fineweb-16k", split="train").with_format("torch")
```

```python
# We set up a Tracer object, which is a utility class to find interesting interactions between two SAEs around an MLP.
# Let's inspect a middle layer.
tracer = Tracerfw(model, layer=7, inp=dict(expansion=4), out=dict(expansion=4))

# We then create a visualizer for both SAEs.
# Implementation-wise, this queries some pre-computed max-activations and shows them in a nice format.
inp_vis = Visualizerfw(model, tracer.inp)
out_vis = Visualizerfw(model, tracer.out)
```

```python
# Compute output features whose top eigenvalues are high, likely indicating some interesting structure.
eigenvals = tracer.compute(compute_truncated_eigenvalues, project=False, k=2)
vals, idxs = eigenvals.topk(15)

# Plot the cosine similarity between these features to see if any are related.
dirs = tracer.out.w_dec.weight[:, idxs]
sims = torch.cosine_similarity(dirs[..., None], dirs[:, None], dim=0)

# Visualize them nicely.
labels = [f"{i}" for i in idxs.cpu()]
px.imshow(sims.cpu(), color_continuous_scale="RdBu", color_continuous_midpoint=0, x=labels, y=labels)
```

```python
import torch
import plotly.express as px

try:
    import numpy as np
    from scipy.cluster.hierarchy import linkage, leaves_list
except ImportError:
    linkage = leaves_list = None


q = tracer.q(1372, project=True)

k = 15
idxs_flat = q.flatten().abs().topk(k * 3).indices
i1, i2 = torch.unravel_index(idxs_flat, q.shape)
idxs = torch.cat([i1, i2]).unique()[:k]

M = q[idxs][:, idxs] 


if linkage is not None:
    M_np = M.detach().cpu().float().numpy()

    A = np.abs(M_np)

    eps = 1e-8
    A0 = A - A.mean(axis=1, keepdims=True)
    denom = (np.linalg.norm(A0, axis=1, keepdims=True) + eps)
    Z = A0 / denom
    corr = Z @ Z.T
    dist = 1.0 - corr

    iu = np.triu_indices_from(dist, k=1)
    dist_condensed = dist[iu]

    L = linkage(dist_condensed, method="average")
    order = leaves_list(L)
    order = torch.tensor(order, device=idxs.device)
else:
    A = M.abs()
    start = int(A.sum(dim=1).argmax().item())
    remaining = set(range(k))
    order_list = [start]
    remaining.remove(start)
    while remaining:
        last = order_list[-1]
        nxt = max(remaining, key=lambda j: float(A[last, j].item()))
        order_list.append(nxt)
        remaining.remove(nxt)
    order = torch.tensor(order_list, device=idxs.device)

idxs_ord = idxs[order]
labels = [f"{i}" for i in idxs_ord.detach().cpu().tolist()]

M_ord = q[idxs_ord][:, idxs_ord].detach().cpu()

fig = px.imshow(
    M_ord,
    color_continuous_midpoint=0,
    color_continuous_scale="RdBu",
    x=labels,
    y=labels,
)
fig.update_xaxes(side="top")
fig.update_yaxes(autorange="reversed")  
fig.show()
```

```python
# Scatter plot: Output feature activation vs Eigenvector-based activation
# For output feature 1372 from layer 7

import numpy as np
from scipy.linalg import eigh
from language import Sight

# Get the bilinear interaction matrix Q for output feature 1372
feature_idx = 1372
Q = tracer.q(feature_idx, project=True)  # Shape: [d_in, d_in]

# Compute top 2 eigenvectors and eigenvalues
Q_np = Q.cpu().numpy()
eigenvalues, eigenvectors = eigh(Q_np)

# Get top 2 eigenvectors (eigh returns in ascending order)
top_2_indices = np.argsort(np.abs(eigenvalues))[-2:]
top_eigenvalues = eigenvalues[top_2_indices]
top_eigenvectors = eigenvectors[:, top_2_indices]

# Reconstruct the rank-2 approximation of Q
Q_approx = np.zeros_like(Q_np)
for i in range(2):
    v = top_eigenvectors[:, i]
    lam = top_eigenvalues[i]
    Q_approx += lam * np.outer(v, v)

Q_approx_torch = torch.tensor(Q_approx, device=device, dtype=Q.dtype)

# Now compute activations on a batch of data
n_samples = 2000  # Number of samples to plot
batch_size = 16

actual_activations = []
approx_activations = []

# Sample from dataset
sample_indices = torch.randperm(len(dataset))[:n_samples]

# Create a fresh Sight object
sight = Sight(model)

for i in range(0, n_samples, batch_size):
    batch_indices = sample_indices[i:i+batch_size]
    input_ids = dataset["input_ids"][batch_indices][:, :model.config.n_ctx].to(device)
    
    with sight.trace(input_ids, validate=False, scan=False):
        # Get input features to the MLP
        x_in = tracer.inp.encode(sight[tracer.inp.point]).save()
        
        # Get actual output feature activation
        x_out = tracer.out.encode(sight[tracer.out.point]).save()
        
    # Get activation of feature 1372
    actual_act = x_out[..., feature_idx]  # [batch, seq]
    
    # Compute approximation outside the trace context
    # For each position: x[pos]^T @ Q_approx @ x[pos]
    batch_size_actual = x_in.shape[0]
    seq_len = x_in.shape[1]
    
    approx_act = torch.zeros(batch_size_actual, seq_len, device=device)
    for b in range(batch_size_actual):
        for s in range(seq_len):
            x_vec = x_in[b, s]  # [d_in]
            approx_act[b, s] = x_vec @ Q_approx_torch @ x_vec
    
    actual_activations.append(actual_act.flatten().cpu())
    approx_activations.append(approx_act.flatten().cpu())

# Concatenate all batches
actual_activations = torch.cat(actual_activations).numpy()
approx_activations = torch.cat(approx_activations).numpy()

# Create scatter plot
fig = px.scatter(
    x=actual_activations,
    y=approx_activations,
    opacity=0.3,
    labels={'x': 'Output feature activation', 'y': 'Eigenvector-based activation'},
    title=f'Output Feature {feature_idx} (Layer {tracer.layer}): Activation vs Top-2 Eigenvector Approximation'
)

# Add diagonal reference line
max_val = max(actual_activations.max(), approx_activations.max())
min_val = 0
fig.add_scatter(
    x=[min_val, max_val],
    y=[min_val, max_val],
    
    mode='lines',
    line=dict(dash='dash', color='orange', width=2),
    showlegend=False
)

fig.update_layout(
    width=600,
    height=600,
    template='plotly_white'
)

fig.show()

print(f"Correlation: {np.corrcoef(actual_activations, approx_activations)[0, 1]:.4f}")
```
